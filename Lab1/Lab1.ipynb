##### Создание нового поля, вычисленного на основе значений других полей

```python
# через выражение на базе имеющихся колонок
df['gdp_per_capita_previous_year'] = 100 *df['gdp_per_capita'] / (100 + df['gdp_variation'])
df
```

```python
# через DataFrame.apply;

def projected_grow_rate(row):
    return row['gdp_per_capita'] + row['gdp_per_capita'] * (0.01 * row['gdp_variation'] - 0.04)

df['projected_gdp_per_capita'] = df.apply(projected_grow_rate, axis=1)
df
```

```python
# через Series.apply.
def desired_growth_rate(value):
    return value * 1.1

df['desired_growth_of_gdp_per_capita'] = df['gdp_per_capita'].apply(desired_growth_rate)
df
```

### 4. Продемонстрировать работу .groupby, на основе группировок в groupby вычислить агрегатные функции по одной или нескольким колонкам

### Решейпинг данных 1D → 2D с использованием .pivot (можно подать на вход результаты агрегатов, полученных ранее с использованием .groupby (сгруппировать по двум полям).

```python
df.pivot(index='country_code', columns='year', values='gdp_per_capita')
```

### Решейпинг 1D → 2D данных, соединённых с группировкой / агрегацией (одним словом - сводная таблица): .pivot_table. Группировать только по категориальным полям или числовым, если значений немного.

```python
bins = [0, 1000, 5000, 10000, 20000, 50000]
df['gdp_bin'] = pd.cut(df['gdp_per_capita'], bins=bins)

df.pivot_table(
    index='country_code',
    columns='gdp_bin',
    values='gdp_per_capita',
    aggfunc='count'
)
```

### Получить DataFrame с MultiIndex любым способом: через конструктор (в документации есть множество видов конструкторов для создания MultiIndex с нуля), через read_sql / read_csv / read_excel, read_*, через pivot_table, через groupby или иными способами.

```python
multi_index_df = df.groupby(['country_code', 'year'])['gdp_per_capita'].mean().unstack()
multi_index_df.columns = pd.MultiIndex.from_product([['GDP'], multi_index_df.columns])
multi_index_df
```

##### Переставить местами уровни индекса

```python
multi_index_df_swapped = multi_index_df.swaplevel(axis='columns')

multi_index_df_swapped
```

##### Транспонировать таблицу (или создать новую другую) с MultiIndex

```python
transposed_df = multi_index_df.T
transposed_df
```

##### Удалить один из уровней индекса или добавить новый уровень индекса (можно инициализированный константой) - см. документацию.

```python
multi_index_df['new_level'] = 'constant_value'
multi_index_df.set_index('new_level', append=True, inplace=True)
multi_index_df
```
